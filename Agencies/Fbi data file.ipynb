## Getting the population from another json

```python
population_2019 = pd.read_csv("./heatmap/states.csv", delimiter="\t")
population_2019.head()
```

```python
json_pop = []

for index, row in population_2019.iterrows():
    string = {"state": row["state"], "latitude": row["latitude"], "longitude": row['longitude'], "name": row['name'], "population": 0}
    json_pop.append(string)
    
    
print(json_pop[:5])
print(len(json_pop))
```

```python
with open('population_.json', 'w') as fout:
    json.dump(json_pop, fout)
```

```python
pop_2019 = pd.read_json("./population_2019.json")
pop_2019.head()
```

```python
df_temp = pop_2019.sort_values('population', ascending=False)
df_temp.plot(kind="bar", x="name", y="population", figsize=(15, 6))
plt.show()
```
