# ETL USA Spending Gov API

This is a public API with access to USA Data Act Government data sets with Federal spending data [USA Spending Gov API](https://api.usaspending.gov/).

**Endpoints spending explorer:**

* Returns spending data information through various types and filters

**Endpoints spending over time:**

* Returns transactions aggregated amounts fro spending over time data visualizations on USAspending.gov

**Endpoints spending by geography:**

* Returns spending by state code, county code, or congressional district code

**Endpoints spending by category:**

* Returns data that is grouped in preset units to support the spending by state territory data visualizations on USAspending.govs advanced search page

**Endpoints state:**

* Returns basic information about the specified state

**Endpoints glossary:**

* Returns a list of glossary terms and definitions

**Endpoints budget functions:**

* Returns all Budget Fucntions associated with a TAS, ordered by Budget Function code

**Endpoints budget subfunctions:**

* Returns all Budget Fucntions associated with a TAS, ordered by Budget Function code

Data has supported formats for SQL and JSON. The call will be for JSON format.

### Extract data from source

```python
# Import dependencies
import requests
import json
```

```python
# URLs
USAspendingUrl = 'https://api.usaspending.gov/'
# Base URLs
spendExplorerUrl = f"{USAspendingUrl}api/v2/spending/"
spendTimeUrl = f"{USAspendingUrl}api/v2/search/spending_over_time/"
spendGeoUrl = f"{USAspendingUrl}api/v2/search/spending_by_geography/"
spendCatUrl = f"{USAspendingUrl}api/v2/search/spending_by_category/"
spendStateUrl = f"{USAspendingUrl}api/v2/recipient/state/"
spendGlossaryUrl = f"{USAspendingUrl}api/v2/references/glossary/"
spendBudgetfunUrl = f"{USAspendingUrl}api/v2/budget_functions/list_budget_functions/"
spendBudgetsubfunUrl = f"{USAspendingUrl}api/v2/budget_functions/list_budget_subfunctions/"
# Retrieve data and convert to json
USAspendExplorer = requests.get(spendExplorerUrl).json()
USAspendTimeUrl = requests.get(spendTimeUrl).json()
USAspendGeoUrl = requests.get(spendGeoUrl).json()
USAspendCatUrl = requests.get(spendCatUrl).json()
USAspendStateUrl = requests.get(spendStateUrl).json()
USAspendGlossaryUrl = requests.get(spendGlossaryUrl).json()
USAspendBudgetfunUrl = requests.get(spendBudgetfunUrl).json()
USAspendBudgetsubfunUrl = requests.get(spendBudgetsubfunUrl).json()
```
